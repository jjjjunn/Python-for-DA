### 4.1.6 팬시 색인
- 팬시 색인(fancy indexing): 정수 배열을 사용한 색인을 설명하기 위해 넘파이에서 차용한 단어
- 리스트나 ndarray로 인덱스 집합을 지정하면 해당 위치의 인덱스에 해당하는 ndarray를 반환

```python
# 0으로 만든 8*4 배열
arr = np.zeros((8,4))
print(arr)

# 해당 배열 값 변경
for i in range(8):
    arr[i] = i
arr
```

```python
# 4번, 3번, 0번, 6번 행 출력
arr[[4,3,0,6]]
```

```python
# 위는 아래와 동일하다
print(arr[4])
print(arr[3])
print(arr[0])
print(arr[6])
```

```python
# -가 붙으면 뒤부터
arr[[-3, -5, -7]]
```

```python
# 다차원 팬시 색인(8행 4열)
arr2 = np.arange(32).reshape((8, 4))
arr2
```

```python
arr2[[1,5,7,2,],[0,3,1,2]] # 1행의 0열, 5행의 3열, 7행의 1열, 2행의 2열
```

```python
print(arr2[[1,3,5],[0,1,2]])
# print(arr2[[1,3,5], [:2]]) # 팬시 인덱싱에서 슬라이싱은 안 됨
print(arr2[[1,3,5],[range(3)]]) # range 사용 가능
```

```python
print(arr2)
row = np.array([0, 2])
print(row)
col = np.array([1, 2])
print(col)

# arr2에 row와 col 적용
print(arr2[row, col])
print(arr2[row, :]) # 0행, 2행, 모든 열 (0,0)(0,1)(0,2)(0,3),(2,0)(2,1)(2,2)(2,3)
print(arr2[:, col]) # 모든 행, 1열, 2열
print(arr2[row, 1]) # 0행 1열, 2행 1열 (0,1)(2,1)
print(arr2[2, col]) # 2행 1열, 2행 2열 (2,1)(2,2)
print(arr2[row, 1:]) # 0행 1-3열, 2행 1-3열 (0,1)(0,2)(0,3),(2,1)(2,2)(2,3)
print(arr2[1:, col]) # 1-7행의 1,2열
```

```python
# 팬시 색인에 값 대입하기 (팬시 색인은 새로운 배열로 복사)
arr2 = np.arange(32).reshape((8, 4))
print(arr2)
print(f"대입하게 될 곳: {arr2[[1,5,7,2,],[0,3,1,2]]}")
arr2[[1,5,7,2,],[0,3,1,2]] = 0
print(f"\n값 대입 후:\n{arr2}")
```

## 4.3 유니버셜 함수
### 단항 유니버셜 함수(1개의 원소)
- abs: 절대값
- sqrt: 제곱근
- square: 제곱
- exp: 지수
- log: 로그 (log10: 로그 10, log2: 로그 2, log1p: 로그 (1+x)
- sign: 원소의 부호 (1-양수, 0-영, -1-음수)
- ceil: 소수점 올림
- floor: 소수점 내림 (dtype 유지)
- rint: 소수점 반올림 (dtype 유지)
- modf: 원소의 몫과 나머지 (배열로 반환)
- isnan: 원소의 숫자 여부 출력(불리언 타입)
- isfinite(isinf): 원소의 유한 여부 불리얼 배열 반환
- cos, cosh, sin, sinh, tan, tanh: 삼각함수, 쌍곡선삼각함수 (코사인, 사인, 탄젠트)

### 이항 유니버셜 함수 (2개의 원소)
- add: 두 배열에서 같은 위치 원소끼리 덧셈
- subrtact: 첫 번째 배열 원소 - 두 번째 배열 원소 (뺄셈)
- multiply: 배열의 원소끼리 곱셈
- divide: 첫 번째 배열 원소 / 두 번째 배열 원소 (나눗셈)
- floor_divide: 첫 번째 배열 원소 / 두 번째 배열 원소 (몫만 취함)
- power: 첫 번째 배열 원소를 두 번째 배열의 원소만큼 제곱
- maximum: 각 배열의 두 원소 중 큰 값 반환 (fmax: NaN 값 무시)
- minumum: 각 배열의 두 원소 중 작은 값 반환 (fmin: NaN 값 무시)
- mod: 첫 번째 배열의 원소를 두 번째 배열의 원소로 나눈 나머지
- copysign: 첫 번째 배열의 원소를 두 번째 배열의 원소의 기호로 바꿈
- greater, greater_equal, less, less_equal, equal, not_equal:
<br> 두 원소 간의 >,>=,<,<=,=,!= 비교 연산 결과 불리언 값 반환
- logical_and: 두 원소 간의 AND)&) 논리 연산 결과 반환
- logical_or: 두 원소 간의 OR(|) 논리 연산 결과 반환
- logical_xor: 두 원소 간의 XOR(^) 논리 연산 결과 반환

```python
arr6 = np.arange(10)
print(arr6)
# sqrt: 제곱근
print(np.sqrt(arr6))
```

### 4.4 배열을 이용한 배열 기반 프로그래밍
- 벡터화: 배열 연산을 사용해서 반복문을 명시적으로 제거하는 기법

```python
points = np.arange(-5, 5, 0.01) # -5부터 4.99가지 0.01씩 증가하는 값 배열
xs, ys = np.meshgrid(points, points) # 두 개의 1차원 배열을 받아 모든 (x, y) 짝을 만듦.
# 그리드 상의 두 점을 가지고 계산
z = np.sqrt(xs **2 + ys **2)
print(z)
```

## 4.4.1 배열 연산으로 조건부 표현하기
- numpy.where: x if 조건 else y 같은 삼항식의 벡터화된 버전

```python
# numpy.where

xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])

cond = np.array([True, False, True, True, False])

result = np.where(cond, xarr, yarr) # cond가 참일 때 xarr 값, 거짓일 때 yarr 값 대입
result1 = np.where(cond, yarr, xarr) # cond가 참일 때 yarr 값, 거짓일 때 xarr 값 대입
print(result)
print(result1)
```

```python
arr8 = rng.standard_normal((4,4))
arr8
```

```python
arr8 > 0
```

### 4.4.2 수학 메서드와 통계 메서드
- sum: 합계
- mean: 평균
- std: 표준 편차
- var: 분산
- min: 최대, max: 최소
- argmin: 최솟값의 인덱스(위치), argmax: 최댓값의 인덱스
- cumsum: 누적합
- cumprod: 누적곱


```python
arr9 = np.arange(1,7).reshape(2,-1)
arr9
```

```python
arr9.mean() # 평균
```

```python
np.mean(arr9)
```

```python
# mean과 sum은 axis 인수를 받아 계산할 수 있다. (0:행 1:열)
print(arr9.sum(axis=0))
print(arr9.sum(axis=1))
print(arr9.mean(axis=0))
print(arr9.mean(axis=1))
```

```python
np.cumsum(arr9) # 누적곱
```

```python
np.cumsum(arr9, axis=0) # 행방향 누적합
```

```python
np.cumsum(arr9, axis=1) # 열방향 누적합
```

```python
np.cumprod(arr9) # 누적곱
```

```python
np.cumprod(arr9, axis=0) # 행방향 누적곱
```

```python
np.cumprod(arr9, axis=1) # 열방향 누적곱
```

## 4.5 배열 데이터의 파일 입출력
- np.save(): 저장
- np.load(): 로드 * 확장자는 .npy
- np.savez(): 여러 개의 배열을 압축된 형식으로 저장  * 확장자는 .npz
- np.savez_compressed: 압축이 잘되는 형식의 데이터 저장 * 확장자는 .npz

```python
arr12 = np.arange(10)
np.save("some_array", arr12)
```

```python
np.load("some_array.npy")
```

```python
np.savez("array_archive.npz", a=arr12, b=arr12)
```

```python
arch = np.load("array_archive.npz")
arch["b"]
```

```python
np.savez_compressed("array_compressed.npz", a=arr12, b=arr12)
```

```python
arch = np.load("array_compressed.npz")
arch["b"]
```

### 4.7 계단오르내리기

```python
position = 0
walk = [position]
nsteps = 1000
for _ in range(nsteps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
plt.plot(walk)
```

```python
nsteps = 1000
rng = np.random.default_rng(seed=12345) # 새로운 난수 생성기
draws = rng.integers(0, 2, size=nsteps)
steps = np.where(draws == 0, 1, -1)
walk = steps.cumsum()
print(f"누적합: {walk}")

print(f"가장 내려간 위치: {walk.min()}") # 최소 위치
print(f"가장 올라간 위치: {walk.max()}") # 최대 위치
print(f"최초로 10칸 떨어진 위치: {(np.abs(walk) >= 10).argmax()}")
```
